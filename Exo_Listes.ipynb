```python
#Exercice 1 : Jours de la semaine

#Constituez une liste semaine contenant les 7 jours de la semaine.
#1. À partir de cette liste, comment récupérez-vous seulement les 5 premiers jours de la semaine d’une part, et ceux du week-end d’autre part ? Utilisez pour cela l’indexation positive.
#2. Cherchez un autre moyen pour arriver au même résultat (en utilisant d'indexation négative).
#3. Trouvez deux manières pour accéder au dernier jour de la semaine.
#4. Inversez les jours de la semaine en une commande

Jours_semaines = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']
Jours = Jours_semaines[:5]
weeckend = Jours_semaines[-2:]
print(Jours_semaines)
print(Jours)
print(weeckend)

#3. Trouvez deux manières pour accéder au dernier jour de la semaine.
Last_jour = Jours_semaines[-1]
print(Last_jour)
#4. Inversez les jours de la semaine en une commande
Jours_semaines.reverse()
#print(Inv_jours)
print(Jours_semaines)
```

```python
#Exercice 2 : Saisons

#Créez 4 listes hiver, printemps, ete et automne contenant les mois correspondants à ces saisons. Créez ensuite une liste saisons contenant les listes hiver, printemps, ete et automne. Vérifiez les instructions suivantes :
#1. saisons[2]
#2. saisons[1][0]
#3. saisons[1:2]
#4. saisons[:][1]. Comment expliquez-vous ce dernier résultat ?

hiver =['Decembre', 'Janvier', 'Fevrier']
Printemps = ['Mars', 'Avril', 'Mai']
ete = ['Juin', 'Juillet', 'Aout']
automne = ['Septembre', 'Octobre', 'Novembre']
Saisons = [hiver, Printemps, ete, automne]

print(Saisons)
F = Saisons[2]
print(F)
S = Saisons[1][0]
print(S)
T = Saisons[1:2]
print(T)
Fo = Saisons[:][1]
print(Fo)
```
