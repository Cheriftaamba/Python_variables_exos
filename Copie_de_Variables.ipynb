### <font color ='bisque'> Operations sur les variables numériques

```python

# somme x avec une valeur numérique
x1 = 5
x2 = 3
x = x1 + x2
# print(x)
# soustraire x avec une valeur numérique
y = x1- x2
# print(y)
# multiplier x avec une valeur numérique
z = x1*x2
# print(z)
# soustraire deux variables numériques

# plusieurs opérations en même temps
t = x1*x2 + (x1- x2)
print(t)
```

```python
i = 2
# incrémentation par sommation (+=)
i += 1 # i = i +1
# print(i)
# incrémentation par soubstrastion (-=)
i -= 3 # i  = i - 3
# print(i)
# incrémentation par multiplication (*=)
k = 5
k *=2  # k = k * 2
# print(k)
# incrémentation par division (/=)
k /= 2 # k = k /2
print(k)
```
